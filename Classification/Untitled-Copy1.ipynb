```python
import numpy as np
import sys
```

```python
file = open("c_no_hurry.in", "r") 
vett=file.readlines()
file.close()
def removeEnter(vett):
    for i in range(len(vett)):
        vett[i]=vett[i].replace("\n","")
    return vett
vett=removeEnter(vett)
vett
```

```python
firstRow=vett[0].split(" ")
rows=int(firstRow[0])
columns=int(firstRow[1])
cars=int(firstRow[2])
rides=int(firstRow[3])
bonus=int(firstRow[4])
steps=int(firstRow[5])
print("First row = ",firstRow)
```

```python
data=np.empty([len(vett)-1,6])
for i in range(1,len(vett)):
    data[i-1]=vett[i].split(" ")
```

```python
def distanceTwoPoint(r1,c1,r2,c2):
    dist=abs(r2-r1)+abs(c2-c1)
    return abs(dist)
startEndDist=np.empty([rides])
for i in range(rides):
    startEndDist[i]=distanceTwoPoint(data[i,0],data[i,1],data[i,2],data[i,3])
```

```python
carsProp=np.zeros([cars,3])#X, Y, delay
```

```python
carRide=[""]*cars
```

### A ciascuna corsa devo assegnare un'auto
Assegno una corsa ad un auto, aggiorno la posizione di quell'auto, ricalcolo le distanze (sommando a ciascuna auto usata il tempo in più che ha già usato) e seleziono l'auto con la distanza minore da un punto

```python
def distanceCarsStartRide(n):
    distances=np.zeros([cars])
    for i in range(cars):
        distances[i]=distanceTwoPoint(data[n,0],data[n,1],carsProp[i,0],carsProp[i,1])+carsProp[i,2]
    return distances

def selectBestChoice(n):
    distances=distanceCarsStartRide(n)
    bestIndex=np.argmin(distances)
    carRide[bestIndex]=str(carRide[bestIndex])+" "+str(n)
    #Change proprieties of the car I used
    carsProp[bestIndex,0]=data[bestIndex,2]
    carsProp[bestIndex,1]=data[bestIndex,3]
    i=bestIndex
    carsProp[i,2]=carsProp[i,2]+startEndDist[n]+data[n,4]+distanceTwoPoint(carsProp[i,0],carsProp[i,1],data[n,0],data[n,1])

def distFromOrigin():
    distFromOrigin=np.zeros([rides])
    for i in range(rides):
        wait=data[i,4]-(data[i,0]+data[i,1])
        if wait>0:
            distFromOrigin[i]=data[i,0]+data[i,1]+wait
        else:
            distFromOrigin[i]=data[i,0]+data[i,1]
    return distFromOrigin
```

```python
def bestCarForRide(n):
    best=sys.maxsize
    bestIndex=0
    for i in range(cars):
        dist=distanceTwoPoint(carsProp[i,0],carsProp[i,1],data[n,0],data[n,1])+carsProp[i,2]+startEndDist[n]
        if(dist<best):
            best=dist
            bestIndex=i
    return best
            
def distancesCarToRide(v):
    distances=np.zeros([rides])
    positive=v[v>=0]
    for i in positive:
        distances[i]=bestCarForRide(i)
    distances[distances==0]=sys.maxsize
    return distances
```

```python
v=np.arange(rides)
distances=distancesCarToRide(v)
whenStart=data[:,4]
whenFinish=data[:,5]
distOrigin=distFromOrigin()
for i in range(rides):
    #distances=distancesCarToRide(v)
    bestRide=np.argmin(whenStart)
    selectBestChoice(bestRide)
    whenStart[bestRide]=sys.maxsize
    #v[bestRide]=-1
```
